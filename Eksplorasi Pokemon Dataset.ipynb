# Analisa Type 1 Pada Pokemon Dataset

```python
print('Berikut adalah beberapa tipe pokemon yang ada dalam kolom Type 1 : \n',df['Type 1'].unique(),'\n\n')
print('Total tipe pokemon yang ada di kolom Type 1 adalah : \n',df['Type 1'].nunique(),'tipe pokemon yang berbeda \n\n')
print('Berikut daftar tipe pokemon dalam kolom Type 1 : \n',df['Type 1'].value_counts())
```

```python
type1_count = df['Type 1'].value_counts()
fig,ax = plt.subplots(figsize=(18,8))

for i in range(type1_count.shape[0]):
        ax.text(i, type1_count.values[i]+1, type1_count.values[i], horizontalalignment='center')
        
ax.bar(type1_count.index, type1_count.values, color='red')
ax.set_xlabel('Tipe pokemon')
ax.set_ylabel('Angka dari Pokemon Type1 1')
ax.set_title('Jumlah Pokemon Type 1')


plt.grid()
```

- Water menjadi tipe pokemon terbanyak pada type 1
- Flying merupakan tipe pokemon paling sedikit pada type 1

# Perbandingan Jumlah Tipe Pokemon Antara Type 1 Dan Type 2

```python
type1_count = df['Type 1'].value_counts()
type2_count = df['Type 2'].value_counts()
fig,ax = plt.subplots(figsize=(18,8))
ax.bar(type1_count.index, type1_count.values, color='red', label='Type 1')
ax.bar(type2_count.index, type2_count.values, color='green', label='Type ')
ax.set_xlabel('Tipe pokemon')
ax.set_ylabel('Total Angka dari Tipe Pokemon')
ax.set_title('Perbandingan Jumlah Tipe Pokemon Antara Type 1 dan Type 2')
plt.legend()
plt.grid()    
```

- Rata-rata type 1 memiliki jumlah yang lebih banyak dibandingkan dengan type 2, hal ini bisa disebabkan juga karena banyak data pada type 2 yang null, berarti pokemon hanya memiliki satu tipe saja.
